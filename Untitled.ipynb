```python
import requests
from bs4 import BeautifulSoup
import pandas as pd 
import numpy as np
from time import sleep
```

```python
url = "https://www.gsmarena.com/makers.php3"

response = requests.get(url)

soup = BeautifulSoup(response.text, "html.parser")
```

```python
print(soup)
```

```python
prueba=soup.find_all("td")
prueba
```

```python
enlace = prueba["href"]
enlace
```

```python
td_list = soup.find_all("td")

urls = []

for td in td_list:
    a_element = td.find("a")

    if a_element:
        url = a_element.get('href')
        urls.append(url)

urls = list(set(urls))

print(urls)
```

```python
url_movil=[]
for i in urls:
    url_moviles =  f"https://www.gsmarena.com/{i}"
    url_movil.append(url_moviles)
url_movil
```
