```python
import pandas as pd 
import numpy as np 
import time 
import requests

```

```python
url_base = "https://api.rawg.io/api"
response = requests.get(f"{url_base}/games?key=5b71558fe711441385839c10cc92fb84")
data = response.json()
```

```python
df = pd.DataFrame(data["results"])
```

```python
for i in range(2, data["count"] // 20 + 2):
    response = requests.get(f"{url_base}/games?key=5b71558fe711441385839c10cc92fb84&page={i}")
    data = response.json()
    df = pd.concat([df, pd.DataFrame(data["results"])], ignore_index=True)
```
